# Task 0 - BMI calculation

First, write code to capture the height and weight entered by the user, then calculate the BMI value and determine the BMI category.The BMI calculation formula is :


*  BMI = weight (kg) / [height (m)]2


Then determine the user's BMI category based on the calculated BMI value:


*   Underweight: BMI < 18.5
*   Normal weight: 18.5 ≤ BMI < 24.9
*   Overweight: 25 ≤ BMI < 29.9
*   Obesity: BMI ≥ 30

BMI.png






```python
# FUNCTION:calculate_bmi
# create function to calculate bmi

def calculate_bmi(weight, height):

    #calculate the bmi
    bmi = weight / (height ** 2)

    #bmi retains two decimal places
    bmi = round(bmi, 2)

    #print bmi
    return bmi

# create function to determine the BMI category based on the BMI value
def bmi_category(bmi):

    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 24.9:
        return 'Normal weight'
    elif 25 <= bmi < 29.9:
        return 'Overweight'
    else:
        return 'Obesity'
```

# Task 2 - Get user input

Get the user's age, gender, weight, height, smoking status, alcohol_status, and activity_status data.

```python

#Get user input

# get user input - name
name = input("Please enter name: ")

# get user input - age
age = int(input("Please enter your age: "))

# get user input - gender
gender = input("Please enter your gender (Male, Female): ")

# get user input - weight
weight = float(input("Please enter your weight (kg): "))

# get user input - height
height = float(input("Please enter your height (m): "))

# get user input - smoking_status
smoking_status = input("Please enter your smoking status (Smoker, Non-smoker): ")

# get user input - alcohol_status
alcohol_status = input("Please enter your alcohol consumption status (Drinker, Non-drinker): ")

# get user input - activity_status
activity_status = input("Please enter your physical activity status (Active, Inactive): ")
```

# Task 3 - Prediction results

Determine the user's heart disease risk prediction results based on the health information entered by the user.

```python
# calculate BMI
bmi = calculate_bmi(weight, height)
# determine BMI category
bmi_cat = bmi_category(bmi)


#Predict the heart condition risk based on the user inputs
risk = risk_matrix.get((bmi_cat, smoking_status, alcohol_status, activity_status), 'Unknown')

# print out the user-entered health information
#use table to display the input results
!pip install tabulate

from tabulate import tabulate

data = [
    ["Name", name],
    ["Age", age],
    ["Gender", gender],
    ["Weight", weight],
    ["BMI Category", bmi_cat],
    ["Smoking Status", smoking_status],
    ["Alcohol Status", alcohol_status],
    ["Activity Status", activity_status],
    ["\033[1;32mHeart Condition Risk\033[0m", f"\033[1;32m{risk}\033[0m"] #Change the color of Heart Condition Risk
]

print(tabulate(data, headers=["Field", "Value"], tablefmt="double_grid"))

```

# Task 4 -Store data in CSV file

Store the user input data and prediction results into a CSV file for later viewing or analysis.

```python
#import csv module
import os
import csv
file_path = 'heart_condition_risk_data.csv'

file_is_empty = not os.path.exists(file_path) or os.stat(file_path).st_size == 0

# Open file and write records
with open(file_path, mode='a', newline='') as file:
    writer = csv.writer(file)

    if file_is_empty:
        writer.writerow(['Name', 'Age', 'Gender', 'Weight', 'Height', 'BMI', 'BMI Category', 'Smoking Status', 'Alcohol Status', 'Activity Status', 'Risk'])


    # continue for more data
    writer.writerow([name, age, gender, weight, height, bmi, bmi_cat, smoking_status, alcohol_status, activity_status, risk])

print("Your data has been saved to the CSV file.")
```
